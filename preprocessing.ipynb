```python
# UYGULAMAYERLERİ ve TANİLAR DEĞİŞKENLERİNİN EKSİK DEĞERLERLERİNİ KNNIMPUTER İLE DOLDURMAK İÇİN KATEGORİK DEĞİŞKENLERE ENCODER İŞLEMLERİNİN UYGULANMASI
```

```python
df["TedaviSuresi"] = df["TedaviSuresi"].str.replace("Seans", "").str.strip().astype(int)
df["UygulamaSuresi"] = df["UygulamaSuresi"].str.replace("Dakika", "").str.strip().astype(int)
```

```python
df["Cinsiyet"] = df["Cinsiyet"].map({"Kadın": 1, "Erkek":0})
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

az_unique_cols = ["KanGrubu", "Uyruk", "Bolum"]

ohe = OneHotEncoder(sparse_output=False, drop="first")
ohe_result = ohe.fit_transform(df[az_unique_cols])

ohe_df = pd.DataFrame(ohe_result, columns=ohe.get_feature_names_out(az_unique_cols))
df = pd.concat([df.reset_index(drop=True), ohe_df], axis=1)
df.drop(columns=az_unique_cols, inplace=True)

cok_unique_cols = ['KronikHastalik', 'Tanilar', 'TedaviAdi', 'UygulamaYerleri', "Alerji"]

for col in cok_unique_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
from sklearn.impute import KNNImputer

impute_cols = ["Tanilar", "UygulamaYerleri"]

knn_features = df.drop(columns=["HastaNo"]).columns.tolist()

imputer = KNNImputer(n_neighbors=5)

df_knn_array = imputer.fit_transform(df[knn_features])

df_knn = pd.DataFrame(df_knn_array, columns=knn_features)

df[impute_cols] = df_knn[impute_cols]
```

```python
# SAYISAL DEĞİŞKENLERİN NORMALLEŞTİRİLMESİ VE STANDARTLAŞTIRILMASI
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_age = StandardScaler()

df["Yas"] = scaler_age.fit_transform(df[["Yas"]])

scaler_minmax = MinMaxScaler()
df[["TedaviSuresi", "UygulamaSuressi"]] = scaler_minmax.fit_transform(df[["TedaviSuresi", "UygulamaSuresi"]])
```
